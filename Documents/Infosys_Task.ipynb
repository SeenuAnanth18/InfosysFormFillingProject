### **Inheritance:**
In Python, inheritance allows a new class (called a child or subclass) to inherit attributes and methods from an existing class (called a parent or superclass). This means the child class can use the methods and properties of the parent class while also adding its own unique features.\
**Example:**


```python
# Parent class
class Animal:
    def speak(self):
        print("Animal speaks")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating an object of Dog
dog = Dog()
dog.speak()  # Inherited from Animal class
dog.bark()   # Defined in Dog class
```

**Explanation :** \
In this example, the Dog class inherits the speak() method from the Animal class but also defines its own bark() method. This allows the Dog class to reuse functionality from the Animal class while adding its own behavior.\
### **Encapsulation :**
Encapsulation is the bundling of data (attributes) and methods (functions) that operate on the data into a single unit (class) while restricting direct access to the data. It protects the internal state of an object and ensures controlled access through public methods (getters and setters).\
**Example:**


```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number  # Private attribute
        self.__balance = balance  # Private attribute

    # Getter for balance
    def get_balance(self):
        return self.__balance

    # Setter for balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Invalid deposit amount!")

# Creating an object
account = BankAccount("12345", 1000)

# Accessing balance using getter
print(account.get_balance())  # Output: 1000

# Modifying balance using setter
account.deposit(500)
print(account.get_balance())  # Output: 1500
```

**Explanation:**\
The __balance attribute is private, and you cannot access it directly (account.__balance). Instead, we use the public methods get_balance and deposit to interact with the balance. This is encapsulation in action.\
### **Abstraction:**
Abstraction hides the internal implementation details of a class and only shows the essential features to the user. It is achieved using abstract classes and methods.\
**Example:**


```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

# Using subclasses
rect = Rectangle(10, 5)
circle = Circle(7)

print(rect.area())  # Output: 50
print(circle.area())  # Output: 153.86

```

**Explanation:**\
The Shape class defines an abstract method area, but does not provide its implementation. Subclasses Rectangle and Circle implement the area method in their specific ways. Users of the class only interact with the area method and are not concerned with how it is calculated.\
### **Polymorphism :**
Polymorphism allows methods in different classes to have the same name but behave differently based on the object that calls them.\
**Example:**


```python
class Bird:
    def sound(self):
        print("Bird makes sound")

class Sparrow(Bird):
    def sound(self):
        print("Sparrow chirps")

class Parrot(Bird):
    def sound(self):
        print("Parrot talks")

# Using polymorphism
def make_sound(bird):
    bird.sound()

sparrow = Sparrow()
parrot = Parrot()

make_sound(sparrow)  # Output: Sparrow chirps
make_sound(parrot)   # Output: Parrot talks
```

**Explanation:**\
The sound method is present in both Sparrow and Parrot classes but behaves differently. When make_sound is called, Python determines which sound method to execute based on the object passed, demonstrating polymorphism.\

## **Lambda Function :**
A lambda in Python is like a shortcut for writing small, quick functions without giving them a name. I find it especially useful when I need a simple function for a specific task without the need to define it with def. It helps make my code cleaner and more concise, especially in cases like filtering or transforming data.
### **1. map()**
The map() function applies a specific operation to every item in a sequence. From my perspective, its like automating repetitive tasks.\
**Example:**\
Imagine I have a list of prices, and I need to add GST (say 18%) to each price.


```python
prices = [100, 200, 300]
add_gst = map(lambda price: price * 1.18, prices)
print(list(add_gst))  # Output: [118.0, 236.0, 354.0]
```

Instead of manually calculating GST for each item, I use map() to handle all prices at once efficiently.
### **2. filter()**
The filter() function selects items from a sequence based on a condition. Its like sorting things I need and ignoring the rest.\
**Example:** \
Imagine I have a list of ages and need to filter only those eligible to vote (18+).


```python
ages = [15, 22, 17, 19, 34]
eligible_voters = filter(lambda age: age >= 18, ages)
print(list(eligible_voters))  # Output: [22, 19, 34]
```

This saves me time by filtering out irrelevant items in one step, like separating eligible voters from a crowd.
### **3. reduce()**
The reduce() function combines all items in a sequence into a single value. Its like summing or multiplying everything together in one go.\
**Example:** \
Imagine I need to find the total sales from a list of daily sales figures.


```python
from functools import reduce
sales = [100, 200, 300]
total_sales = reduce(lambda x, y: x + y, sales)
print(total_sales)  # Output: 600
```

Its a simple way to aggregate data, like calculating total revenue without writing loops.
### **4. List Comprehensions**
These are an alternative to map() and filter() for creating lists in a cleaner, more readable way.\
**Example:** \
Suppose I want to convert all product names to uppercase.


```python
products = ["laptop", "phone", "tablet"]
uppercase_products = [product.upper() for product in products]
print(uppercase_products)  # Output: ['LAPTOP', 'PHONE', 'TABLET']
```

This feels more intuitive and readable compared to other methods, especially for straightforward operations.
### **5. zip()**
The zip() function combines multiple lists into pairs.\
**Example:** \
Imagine I have two lists: names of employees and their salaries, and I want to pair them.


```python
names = ["Alice", "Bob", "Charlie"]
salaries = [50000, 60000, 70000]
employee_data = zip(names, salaries)
print(list(employee_data))  # Output: [('Alice', 50000), ('Bob', 60000), ('Charlie', 70000)]
```

This makes it easy to organize related data into a structured format, like combining employee details.

## **Loops :**
For loops and while loops are fundamental in Python, and they allow me to handle repetitive tasks efficiently. Each has its own purpose, depending on whether the number of iterations is predetermined or dynamic.
### **For Loop:**
A for loop is used when I know how many iterations are needed, or when Im working with a sequence like a list, string, or range. It simplifies repetitive tasks and is highly structured.\
**Example 1: Sending Bulk Emails**\
Imagine Im tasked with sending promotional emails to a list of customers.


```python
customers = ["Alice", "Bob", "Charlie"]
for customer in customers:
    print(f"Sending email to {customer}")
```

**Explanation:**\
Here, the for loop goes through each customer in the list and sends a personalized email. This avoids manually writing separate emails for each person.\
**Example 2: Grocery Billing System**\
In a grocery store, I might need to calculate the total cost of items purchased.


```python
prices = [100, 200, 150, 50]
total = 0
for price in prices:
    total += price
print(f"The total bill is: {total}")
```

**Explanation:**\
The for loop calculates the total cost by summing up the prices of items. This is useful for automating billing in retail or e-commerce systems.\
###**While Loop:**
A while loop is useful when the number of iterations depends on a condition that may change dynamically during execution.\
**Example 1: ATM PIN Validation**\
At an ATM, the system allows users up to three attempts to enter the correct PIN.


```python
attempts = 3
while attempts > 0:
    pin = input("Enter your PIN: ")
    if pin == "1234":
        print("Access Granted!")
        break
    else:
        attempts -= 1
        print(f"Incorrect PIN. {attempts} attempts left.")
if attempts == 0:
    print("Account locked due to too many failed attempts.")
```

**Explanation:**\
The while loop keeps asking for the PIN until the user enters it correctly or runs out of attempts. This real-world example demonstrates how while loops can handle dynamic input and ensure security.\
**Example 2: Ride Booking App**\
In an app like Uber, the system waits for a driver to accept a ride request.


```python
ride_accepted = False
while not ride_accepted:
    user_input = input("Has a driver accepted the ride? (yes/no): ")
    if user_input.lower() == "yes":
        ride_accepted = True
        print("Ride confirmed!")
    else:
        print("Searching for drivers...")
```

**Explanation:**\
The while loop keeps running until a driver accepts the ride. This is a practical way to manage dynamic processes in real-time applications like ride-hailing services.
###**For Loop with Nested Loop:**
A nested loop is when a loop is placed inside another loop. Its helpful for tasks involving multiple layers of repetition.\
**Example 1: Generating a Multiplication Table**\
In a classroom scenario, I may need to generate a multiplication table for numbers from 1 to 5.


```python
for i in range(1, 6):  # Outer loop for numbers 1 to 5
    for j in range(1, 11):  # Inner loop for multiplying 1 to 10
        print(f"{i} x {j} = {i * j}")
    print("----------")
```

**Explanation:**\
The outer loop iterates through numbers 1 to 5, and the inner loop multiplies each number by 1 to 10. This creates a complete multiplication table.\
**Example 2: Seating Arrangement in a Theater**\
In a theater, each row has multiple seats, and I need to label them.


```python
rows = ["A", "B", "C"]
seats = [1, 2, 3, 4]
for row in rows:
    for seat in seats:
        print(f"Seat: {row}{seat}")
```

**Explanation:**\
The outer loop iterates through rows, while the inner loop assigns seat numbers for each row. This is useful for managing seating arrangements in cinemas or events.
### **While Loop with Nested Loop**
Nested while loops can handle scenarios where multiple conditions or levels of iterations are involved.\
**Example 1: ATM Denomination Distribution**\
An ATM dispenses cash in denominations of 2000, 500, and 100. A nested loop can calculate how to distribute the cash for a specific amount.


```python
amount = 3700
denominations = [2000, 500, 100]
index = 0
while amount > 0 and index < len(denominations):
    count = 0
    while amount >= denominations[index]:
        amount -= denominations[index]
        count += 1
    if count > 0:
        print(f"{count} notes of {denominations[index]}")
    index += 1
```

**Explanation:**\
The outer loop iterates through denominations, while the inner loop calculates how many notes of a specific denomination can be given. This ensures an efficient cash distribution process.\
**Example 2: Booking System with Limited Attempts**\
Suppose Im creating a system for booking tickets where each user has three attempts to enter a valid booking code.

```python
users = ["User1", "User2"]
for user in users:
    print(f"Processing booking for {user}")
    attempts = 3
    while attempts > 0:
        code = input("Enter your booking code: ")
        if code == "VALIDCODE":
            print("Booking confirmed!")
            break
        else:
            attempts -= 1
            print(f"Invalid code. {attempts} attempts left.")
    if attempts == 0:
        print("Booking failed for this user.")
```

**Explanation:**\
The outer loop processes multiple users, and the inner while loop ensures each user has three attempts to enter the correct code. This mimics a real-world ticket booking system with limited retry attempts.

Function Description
Complete the timeConversion function in the editor below. It should return a new string representing the input time in 24 hour format.

	string: the time in  hour format

###**My Answer :**


```python
#!/bin/python3

import math
import os
import random
import re
import sys

def timeConversion(s):
    # Extract the hour, minute, and second
    hour = int(s[:2])
    minute_second = s[2:8]
    period = s[-2:]  # AM or PM

    # Convert hour based on AM/PM
    if period == "AM":
        if hour == 12:
            hour = 0  # Midnight case
    elif period == "PM":
        if hour != 12:
            hour += 12  # Convert PM hour to 24-hour format

    # Format the hour to ensure two digits and return the result
    return f"{hour:02}{minute_second}"

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = timeConversion(s)

    fptr.write(result + '\n')

    fptr.close()
```

```python
#!/bin/python3

import os

def caesarCipher(s, k):
    # Initialize an empty list to store the result
    result = []

    # Normalize k to be within 0-25 (since the alphabet has 26 letters)
    k = k % 26

    # Iterate over each character in the string
    for char in s:
        # If the character is lowercase
        if 'a' <= char <= 'z':
            # Shift the character and ensure it wraps around within 'a' to 'z'
            shifted_char = chr(((ord(char) - ord('a') + k) % 26) + ord('a'))
            result.append(shifted_char)
        # If the character is uppercase
        elif 'A' <= char <= 'Z':
            # Shift the character and ensure it wraps around within 'A' to 'Z'
            shifted_char = chr(((ord(char) - ord('A') + k) % 26) + ord('A'))
            result.append(shifted_char)
        else:
            # If it's not a letter, append the character as it is
            result.append(char)

    # Join the list into a string and return it
    return ''.join(result)

if __name__ == '__main__':
    # Open the file to write the result
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    # Read input
    n = int(input().strip())  # length of the string (not needed for the function)
    s = input()  # the unencrypted string
    k = int(input().strip())  # the shift factor

    # Call the caesarCipher function and get the result
    result = caesarCipher(s, k)

    # Write the result to the output file
    fptr.write(result + '\n')

    # Close the file
    fptr.close()
```

```python
import queue as Queue

cntr = 0

class Node:
    def __init__(self, freq, data):
        self.freq = freq
        self.data = data
        self.left = None
        self.right = None
        global cntr
        self._count = cntr
        cntr += 1

    def __lt__(self, other):
        if self.freq != other.freq:
            return self.freq < other.freq
        return self._count < other._count

def huffman_hidden():
    q = Queue.PriorityQueue()

    for key in freq:
        q.put((freq[key], key, Node(freq[key], key)))

    while q.qsize() != 1:
        a = q.get()
        b = q.get()
        obj = Node(a[0] + b[0], '\0')
        obj.left = a[2]
        obj.right = b[2]
        q.put((obj.freq, obj.data, obj))

    root = q.get()
    root = root[2]
    return root

def dfs_hidden(obj, already):
    if obj is None:
        return
    elif obj.data != '\0':
        code_hidden[obj.data] = already

    dfs_hidden(obj.right, already + "1")
    dfs_hidden(obj.left, already + "0")

def decodeHuff(root, s):
    decoded_string = ""
    current_node = root

    for char in s:
        if char == '0':
            current_node = current_node.left
        else:  # char == '1'
            current_node = current_node.right

        if current_node.left is None and current_node.right is None:
            decoded_string += current_node.data
            current_node = root

    print(decoded_string)

# Example Input
ip = input()  # Input string
freq = {}

for ch in ip:
    freq[ch] = freq.get(ch, 0) + 1

root = huffman_hidden()

code_hidden = {}
dfs_hidden(root, "")

if len(code_hidden) == 1:
    for key in code_hidden:
        code_hidden[key] = "0"

toBeDecoded = "".join(code_hidden[ch] for ch in ip)

decodeHuff(root, toBeDecoded)
```

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'noPrefix' function below.
#
# The function accepts STRING_ARRAY words as parameter.
#

def noPrefix(words):
    # Sort the words lexicographically
    words.sort()

    # Iterate through the sorted list of words
    for i in range(len(words) - 1):
        # If the current word is a prefix of the next word
        if words[i + 1].startswith(words[i]):
            print("BAD SET")
            print(words[i])  # Print the prefix word
            return

    # If no prefix conflicts, print GOOD SET
    print("GOOD SET")

if __name__ == '__main__':
    n = int(input().strip())  # Read the number of words

    words = []  # Initialize an empty list for words

    # Read the words and add them to the list
    for _ in range(n):
        words_item = input().strip()  # Read each word
        words.append(words_item)

    # Call the function with the list of words
    noPrefix(words)
```

### **Qn 2: Shape Classes with Area Method**

```python
#!/bin/python3

import math
import os
import random
import re
import sys

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return round(self.length * self.width, 2)

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return round(math.pi * self.radius ** 2, 2)

if __name__ == '__main__':
    q = int(input().strip())  # Number of queries
    for _ in range(q):
        query = input().strip().split()
        shape = query[0]

        if shape == "circle":
            radius = int(query[1])
            circle = Circle(radius)
            print(f"{circle.area():.2f}")

        elif shape == "rectangle":
            length = int(query[1])
            width = int(query[2])
            rectangle = Rectangle(length, width)
            print(f"{rectangle.area():.2f}")
```
